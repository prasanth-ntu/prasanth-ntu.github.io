## Linear Regression 

- Aka ordinary least squares (OLS), is the simplest and most classic linear method. 
- Linear Regression finds the best parameters `w` and `b` that minimises the mean squared error between true output, y and predicted output, h(x) on the training set.
- mean squared error = sum of squared differences between between the predictions and true values

# K-NN Classification

**Meet the data - `forge` dataset from `mglearn`**
- Small and synthetic (i.e. made-up) classification dataset
- Has 2 features 
- Has 2 classes (Class 0 & 1) 
- Has 26 data instances (samples) in total

```python
import mglearn
import matplotlib.pyplot as plt

X,y = mglearn.datasets.make_forge()
print ("X shape:{}".format(X.shape))

# Create a scatter plot to vizualise all data points in the datasset
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.legend(["Class 0", "Class 1"])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

- One of the simplest & most used classification algo.
- It's a non-parametric, lazy learning algo.
    - non-parametric - No parameters are learned from the model
    - lazy algo - There is no explicit training phase involved

## One nearest neighbour

In its simplest version, the `k-NN` algorithm only considers exactly **one nearest neighbor**, *which is the closest training data point to the point we want to make a prediction for* & assigns its label to the test data. 

Figure below illustrates this for the case of classification on the `forge` dataset. Here, 3 new test data points are added (shown in `star`). For each of the test point, the closest training data point is linked. The prediction label of the test point will be that of the closest training point (*as shown by the star color*).

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

## Two nearest neighbours

Instead of considering only the closest neighbor, we can also consider an arbitrary
number, k, of neighbors. **When considering more than one neighbor, we use majority voting to assign a label**. 
- This means that for each test point, we count how many neighbors belong to class 0 and how many neighbors belong to class 1. 
- We then assign the class that is more frequent: in other words, the majority class among the k-nearest neighbors. 

The following figure uses 2 nearest neighours

```python
mglearn.plots.plot_knn_classification(n_neighbors=2)
```

## Three nearest neighbours

We can see that the prediction for the new data point at the top left is not the same as the prediction when we used only one or two neigbours.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
plt.plot(8.2,3.5, 'ro')
```

While this illustration is for a `binary classification` problem, this method can be
applied to datasets with any number of classes `multiclass classification` and of higher dimension in contrast to two-dimensions in this case. For more classes, we count how many neighbors belong to each class and again predict the most common class (`majority voting`).

## K-NN - Training

```python
from sklearn.model_selection import train_test_split
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)

clf.fit(X_train, y_train)

print("Test set predictions: {}".format(clf.predict(X_test)))

print("Test set accuracy: {:.2f}".format(clf.score(X_test, y_test)))
```

## Analyzing K-NN Classifier

### Model complexity vs `n_neighbours`

With two-dimensional datasets, we can easily vizualise the prediction for all possible test points in the xy-plane. Let's color the plane according to the class that would be assigned to a point in this region. This lets us view the `decision boundary`.

<p style="color:green; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;"><b>Decision boundary is the divide between where the algorithm assigns class 0 versus where it assigns class 1 to the new test data.</b></p>


Let's build KNN model for `K=1,3,9` on the `forge` dataset & plot the `decision boundary`.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # build the KNN model
    knn = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    # create the scatter plot with decision boundary
    mglearn.plots.plot_2d_separator(knn, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")
axes[0].legend(loc=3)
fig.suptitle("Decision boundaries created by the nearest neighbors model for different values of n_neighbor", y=1.1, fontsize=14)
plt.show()
```

From the figure on the left, we can see that **using a single neighbor results in a decision boundary that follows the training data closely**. **Considering more and more neighbors leads to a smoother decision boundary**. 

<p style="color:blue; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;">A smoother boundary corresponds to a simpler model. In other words, using few neighbors corresponds to high model com plexity (as shown on the left side of figure with K=1), and using many neighbors corresponds to low model complexity (as shown on the right side of figure with K=9).</p>


<p><b><u>Food for thought!</u></b></p>
*If we consider the extreme case where the number of neighbors is the number of all data points in the training set, which class the test set would belong to!*

### Model complexity vs Generalization

**Meet the data - `Breast Cancer` dataset from `sklearn`**

Let's see the connection between **model complexity** and **generalization** (*ability of the model to make accurate predictions on new, unseen dataset*).  We will do this on the real-world `Breast Cancer` dataset.

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
print (cancer['DESCR'][:1110] + '\n\t...')
```

 We begin by splitting the dataset into a training and a test set.

```python
X_train, X_test, y_train, y_test = train_test_split(cancer.data, 
                                                    cancer.target, 
                                                    stratify=cancer.target, random_state=66)
print ("Training data shape:{}".format(X_train.shape))
print ("Test data shape    :{}".format(X_test.shape))
```

Then we evaluate training and test set performance with different numbers of neighbors.

*Note*: Concepts related to `Generalization, Overfitting & Underfitting` will be covered in a separate post.

```python
training_accuracy = []
test_accuracy = []
# try n_neighbors from 1 to 20
neighbors_settings = range(1, 21)

for n_neighbors in neighbors_settings:
    # intialize the knn model parameters
    knn = KNeighborsClassifier(n_neighbors=n_neighbors)
    # build/ train the model
    knn.fit(X_train, y_train)
    # save training set accuracy
    training_accuracy.append(knn.score(X_train, y_train))
    # save test set (generalization) accuracy
    test_accuracy.append(knn.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, '--', label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.title("Comparison of training and test accuracy as a function of n_neighbors\n")
plt.legend()
plt.show()
```

While real-world plots are rarely very smooth, we can still recognize some of the characteristics of `overfitting` and `underfitting` (note that because considering fewer neighbors corresponds to a more complex model). 
- `overfitting` - Model works well on training set, not on test set as it's not able to generalize.
    - Model is too complex.
- `underfitting` - Model do badly on both training set as well as test set.
    - Model is too simple.

Considering a single nearest neighbor case in the above plot, the prediction accuracy on the training set is perfect (close to 1). But when more neighbors are considered, the model becomes simpler and the training accuracy drops.  The test set accuracy for using a single neighbor is lower than when using more neighbors, indicating that using the single nearest neighbor leads to a model that is too complex. On the other hand, when considering  10 neighbors, the model is too simple and performance is still worse. The worst performance is around 88% test accuracy (for `n_neighbour=2`), which might still be acceptable.

<p style="color:blue; font-size:110%;  border-left: 5px solid blue; padding-left: 10px;">The best performance is somewhere in the middle, using around six neighbors <code>n_neighbour=6</code> with a test accuracy of <code>0.937</code>.</p>

<p style="color:red;">Keep in mind that best <code>n_neighbours</code> vary from dataset to dataset & even within same dataset of different training/ test data size.</p> 

```python
print ('--- Based on test set/ generalization ---\n')

index = test_accuracy.index(max(test_accuracy))
print ('Best n_neighour  = {} | Training accuracy = {:0.3f} | Test accuracy = {:0.3f}'.
       format(neighbors_settings[index], training_accuracy[index], test_accuracy[index]))

index = test_accuracy.index(min(test_accuracy))
print ('Worst n_neighour = {} | Training accuracy = {:0.3f} | Test accuracy = {:0.3f}'.
       format(neighbors_settings[index], training_accuracy[index], test_accuracy[index]))
```

## K-NN Training

```python
from sklearn.neighbors import KNeighborsRegressor

X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print ("Training data shape:{}".format(X_train.shape))
print ("Test data shape    :{}".format(X_test.shape))

# Set the number of neighbors to 3 & fit the KNN model on training set
reg = KNeighborsRegressor(n_neighbors=3).fit(X_train, y_train)
```

```python
# Make predictions on test set
y_predict = reg.predict(X_test)
print("Test set predictions:\n{}".format(y_predict))
```

```python
print ('First row  : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[0], y_predict[0]))
print ('Second row : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[1], y_predict[1]))
print ('Third row  : Target value: {:0.3f} | Predicted value: {:0.3f}'.format(y_test[2], y_predict[2]))
```

<p>For <code>Classificaiton</code> problems, we used <code>accuracy</code> as our evaluation metric. However for regression, we can't use accuracy. There are still many metrics to measure the regression model performance & by default, <code>score</code> method in <code>sklearn</code> returns <code>R<sup>2</sup></code> score.</p>

<p>The <code>R<sup>2</sup></code> score, also known as the <b>coefficient of determination</b>, is a measure of goodness of a prediction for a regression model, and yields a score between 0 and 1. A value of 1 corresponds to a perfect prediction</p>

`R^2 = (1 - u/v)`
where
- `u` is the residual sum of squares `((y_true - y_pred) ** 2).sum()`
- `v` is the total sum of square `((y_true - y_true.mean())**2)`

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

A R^2 score of 0.83 indicates a relatively good model fit.

## Analyzing K-NN Regressor

### Model complexity vs `n_neighbours`

For our one-dimensional `wave` dataset, we can see what the predictions look like for all possible input feature values. To do this, let's create a test dataset consisting of many points on the line:

```python
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle("Comparing predictions made by nearest neighbors regression for dierent values of n_neighbors", y = 1.1, fontsize=14)
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1)
```

```python
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using n_neighbours as 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0), markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1), markersize=8)
    
    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} | test score: {:.2f}".format(
        n_neighbors, reg.score(X_train, y_train), reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target",  "Test data/target"], loc="best")
plt.show()
```

As we can see from the above plot, using only a `single neighbor`, each point in the training set has an obvious influence on the target predictions, and the predicted values go through all of the data points. This leads to a very unsteady prediction. Considering `more neighbors` leads to smoother predictions, but these do not fit the training data as well.

### Model complexity vs Generalization

```python
X, y = mglearn.datasets.make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print ("Training data shape:{}".format(X_train.shape))
print ("Test data shape    :{}".format(X_test.shape))

training_score = []
test_score = []
# try n_neighbors from 1 to 20
neighbors_settings = range(1, 21)

for n_neighbors in neighbors_settings:
    # intialize the knn model parameters
    knn = KNeighborsRegressor(n_neighbors=n_neighbors)
    # build/ train the model
    knn.fit(X_train, y_train)
    # save training set accuracy
    training_score.append(knn.score(X_train, y_train))
    # save test set (generalization) accuracy
    test_score.append(knn.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_score, label="training R^2 score")
plt.plot(neighbors_settings, test_score, '--', label="test R^2 accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.title("Comparison of training and test R^2 score as a function of n_neighbors\n")
plt.legend()
plt.show()
```

# Summary

## Parameters

**The 2 most important parameters of KNN are**: 
- `n_neighbors` - Number of neighbours
- `metric` - how we measure the distance between data points

**`n_neighbors`**
    - Lower `n_neighbors`  - More complex model => Overfitting
    - Higher `n_neighbors` - More simplistic model => Underfitting
    - We should find the sweetspot between a simple & complex model.
We have already seen how to select the best `n_neighbours` based on the test set score (`accuracy or R^2`) (ideally, we should select these parameters using validation set, which will be covered in later posts).

**`metric`**
<p>Default distance <code>metric</code> is <code>minkowski</code> with power parameter of <code>2</code> which generally works well.<br><br>
    
Minkowsi distance of order <code>p</code> between two points = <code>(&#8721;|x<sub>i</sub>y<sub>i</sub>|<sup>p</sup>)<sup>1/p</sup></code> for <code>i=1,...,n</code>
 <ul>
<li><code>minkowski</code> with <code>p=1</code> == <code>manhattan_distance (l1)</code> = <code>(&#8721;|x<sub>i</sub>y<sub>i</sub>|<sup>1</sup>)<sup>1/1</sup></code></li> </li>
    <li><code>minkowski</code> with <code>p=2</code> == <code>euclidean_distance (l2)</code> = <code>(&#8721;|x<sub>i</sub>y<sub>i</sub>|<sup>2</sup>)<sup>1/2</sup></code></li> </li></li>
</ul></p>    
    
<p>Choosing the 'correct' distance metric requires understanding the context. When travelling by plane the <code>Euclidean distance</code> is usually the right notion (ignoring the earths curvature) to approximate travelling time while travelling by cab, the <code>Manhattan metric</code> is a good approximation. So the 'correct' notion of distance depends on what it is that we want to measure.</p>

## Strength & Weakaness

**Strength** 
- Simple & Easy to understand & implement
- Less number of parameters to tune
- Non parametric (i.e. no weights are learnt)
- Fast training phase as it just have to store all the training sampels 
- Works well for multiclass data sets

**Weakness**
- Computationally expensive testing phase (hence, impractical in industrial standards)  
- Prediction can be slow, when the training set is large (interms of number of features or number of samples)
- KNN can suffer from skewed class distributions (certain class being frequent in training set will dominate majority voting). 
- Can't handle higher-dimensional features because there is little difference between the nearest and farthest neighbor.

Data have to be preprocessed (to avoid skewing the distance metric in favor of a particular feature). More details on data `preprocessing` will be covered in a later post.

Using k-NN is a good baseline method before trying more advanced techniques.
